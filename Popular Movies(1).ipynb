```python
 #Number of genres in the dataset
genres_column = df['genres']

unique_genres = set() # create empty set  for not store duplicates genres

for genres_list in genres_column: 
    genres = eval(genres_list) # eval() is used in the code to convert a string representation of a list into an actual list.
    unique_genres.update(genres)  # Add the genres to the set, which automatically removes duplicates
    
len(unique_genres) # it give the length of unique_genres
```

- There are 19 unique genres are present in our dataset.

```python
# To get Top 5 movies from every genres

for genre in unique_genres: # Iterate over each unique genre
    # Filter the DataFrame to get movies that contain the current genre & Sort the filtered movies by popularity in descending order and take the top 10
    top_movies = df[df['genres'].str.contains(genre)].sort_values('popularity', ascending=False).head(5)
    movie_title = top_movies['title']  # get movie title 
    print(genre)
    for i, title in enumerate(movie_title):
       print(i+1, title) 
    print("\n")
```

```python
# To get top 5 production companies with maximum movies count

production_companies = df['production_companies']
unknown='Unknown'

company_count = {} #  Create an empty dictionary to store the count of movies for each production company

# iterate the each rows for companies 
for companies_list in production_companies:
    companies = eval(companies_list)  # Convert the string representation of list to a list
    for company in companies:
        if company in company_count: # # If the company is already in the dictionary, increment its movie count
            company_count[company] += 1  
        else:
            company_count[company] = 1  # other wise it add the initial movie count

# Sort the dictionary items by movie count in descending order
sort_companies = sorted(company_count.items(), key=lambda x: x[1], reverse=True)  # reverse= True it show highest to lowest movie count

# to show top 5 companies
top_5_companies = sort_companies[:5]

for company, count in top_5_companies:
   print(company, ": ", count) # print company name and their count
```

```python
# To find the most popular movie in every 5 year with their popularity

df['release_date']=pd.to_datetime(df['release_date']) # to convert the release date to datetime 

# create a new column for getting moving in every 5 year
df['release_5_year'] = (df['release_date'].dt.year // 5) * 5 # to get 5 year use dt.year

# Group the  'release_5_year' and find the most popular movie in each group nlargest get highest popularity
popular_movie = df.groupby('release_5_year').apply(lambda x: x.nlargest(1, 'popularity')) 

for i, r in popular_movie.iterrows(): # it loop iterate the each rows one by one 
    print(f"Year: {r['release_5_year']}-{r['release_5_year']+4}") # r=row
    print("Movie Name:", r['title'])
    print("Popularity:", r['popularity'])
    print("\n")
```

```python
# calculate Average popularity by year

df['Year']=df['release_date'].dt.year # Create a new column 'Year' by extracting the year from the 'release_date' column

# create a group of popularity by year
avg_popularity=df.groupby('Year')['popularity'].mean() # calculate average using mean()
print(avg_popularity)
```

```python
# Visualization of Rise(increase) and Drop(decrease) of Popularity with different genres over the time

for genre in unique_genres:   # loop iterate through unique_genres
    movies = df[df['genres'].str.contains(genre)]  
    avg_popularity = movies.groupby('Year')['popularity'].mean()  # create a group popularity by year and calculate there average
    plt.plot(avg_popularity.index, avg_popularity.values)
    plt.title('Rise or Drop in Popularity of '+str(genre)+'  Movies over Time')
    plt.xlabel('Year')
    plt.ylabel('Popularity')
    plt.xticks(np.arange(1900, 2030, step=5),rotation=90)
    plt.grid(axis='y')
    plt.show()
```

```python
# Visualization of Rise and Drop of Popularity with different languages over the time

unique_language = {"English","Japanese","French","Chinese","German","Russian"}
   
for language in unique_language:
    movies = df[df['original_language'].str.contains(language)]
    avg_popularity = movies.groupby('Year')['popularity'].mean()
    plt.plot(avg_popularity.index, avg_popularity.values, color='red')
    plt.title('Rise or Drop in Popularity in '+str(language)+' Movies over Time')
    plt.xlabel('Year')
    plt.ylabel('Popularity')
    plt.xticks(np.arange(1900, 2030, step=5),rotation=90)
    plt.grid(axis='y')
    plt.show()
```
